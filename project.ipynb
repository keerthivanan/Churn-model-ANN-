```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle 
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
##preprocess the data 
##drop the irrelevant coloumn
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
#Encode categorical variable
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
label_encoder_gender
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder=OneHotEncoder()
geo_encoder=onehot_encoder.fit_transform(data[['Geography']])
```

```python
geo_encoder 
```

```python
onehot_encoder.get_feature_names_out(['Geography'])
```

```python
geo_encoder.toarray()
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
# combine al the columns to orginal data
data =pd.concat([data.drop('Geography',axis=1),geo_encoder_df],axis=1)
 
```

```python
data.head()
```

```python
##save the enocoders and scaler
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('onehot_encoder.pkl','wb') as file:
    pickle.dump(onehot_encoder,file)
```

```python
##dicide the dataset into indepent and dependent feautures
X=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
##split the data in training and testing sets
X_train,X_test,y_train,Y_test=train_test_split(X,y,train_size=0.2,random_state=42)
```
