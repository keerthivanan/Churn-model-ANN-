```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.tail()
```

```python
##preprocess the data 
##drop the irrelevant coloumn
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
#Encode categorical variable
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder=OneHotEncoder()
geo_encoder=onehot_encoder.fit_transform(data[['Geography']])
```

```python
geo_encoder
```

```python
onehot_encoder.get_feature_names_out(['Geography'])
```

```python
geo_encoder.toarray()
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
# combine al the columns to orginal data
data =pd.concat([data.drop('Geography',axis=1),geo_encoder_df],axis=1)
 
```

```python
data.head()
```

```python
##save the enocoders and scaler
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('onehot_encoder.pkl','wb') as file:
    pickle.dump(onehot_encoder,file)
```

```python
##dicide the dataset into indepent and dependent feautures
X=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
##split the data in training and testing sets
X_train,X_test,y_train,Y_test=train_test_split(X,y,train_size=0.2,random_state=42)
```

ANN implemetantion

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
##building our ann model
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),##HL! connected with input layer
    Dense(32,activation='relu'),##hl2
    Dense(1,activation='sigmoid')##HL2 to output 
]          
)
```

```python
model.summary()
```

```python
X_test.shape[1]
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.1)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
#compile the model
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##set up the tensorboard
log_dir='logs/fit'+datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
##set up early stooping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
##train the model
history=model.fit(

    X_train,y_train,validation_data=(X_test,Y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
